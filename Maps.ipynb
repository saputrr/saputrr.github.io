```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/saputrr/saputrr.github.io/refs/heads/main/IDN_regional_gdp.csv'
df = pd.read_csv(url, skiprows=1)
df.rename(columns={"Unnamed: 0": "province"}, inplace=True)
df = df.iloc[:-2]
df
```

```python
rgdp_df = df.melt(id_vars=['province'], var_name='year', value_name='rgdp') # Moving our data from wide to long format
rgdp_df.columns = ['province', 'year', 'rgdp'] # Renaming the columns

rgdp_df
```

```python
rgdp_df.to_csv('IDN_rgdp.csv', index=False)
```
