```python
import pandas as pd
url = 'https://raw.githubusercontent.com/saputrr/saputrr.github.io/refs/heads/main/API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_9326.csv'
df = pd.read_csv(url, header=4)

df
```

```python
df = df.rename(columns={
    'Country Name': 'country',
    'Country Code':	'code',
    'Indicator Name':	'series',
    'Indicator Code': 'series_code',
})
df.head()
```

```python
df.melt(id_vars=['country', 'code', 'series', 'series_code'])
```

```python
df = df.drop(columns=['series', 'series_code'])
df
```

```python
df1 = df.melt(id_vars=['country', 'code'], var_name='year')

df1
```

```python
df1 = df1.dropna()
df1
```

```python
df1 = df1[df1['country'] == 'Indonesia'].copy()
df1.head()
```

```python
df1.to_csv('GDPpercap.csv', index=False)
```

```python
import altair as alt

chart = alt.Chart(df1).mark_line().encode(
          alt.X('year'),
          alt.Y('value'),
          )
```

```python
chart_json = chart.to_json()
print(chart_json)
```
